## Exploratory Data Analysis On Superstore Dataset 

```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from warnings import filterwarnings
filterwarnings(action='ignore')
```

```python
# Loading Dataset
data = pd.read_csv("SampleSuperstore.csv")
print("Dataset Loaded Successfully")
```

```python
# Data description
data.describe(include="all")
```

```python
plt.figure(figsize=(10,8))
data['Sub-Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(5,8))
data['Region'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(11,11))
data.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.show()
```

```python
heatmap_data = pd.pivot_table(data,values='Profit', index=["City"], columns='Sub-Category')
plt.figure(figsize=(8, 12))
sns.heatmap(heatmap_data)
```
